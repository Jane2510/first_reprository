```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('books (1).csv')

def data_plot(x, y, ax):
    ax.plot(x, y, 'o-', color = 'red')
    ax.set_xlabel('Месяцы')
    ax.set_ylabel('Количество книг')
    ax.set_title('Прочитанные книги за 2023 год')
    ax.legend(title = "Прочитано книг")
    ax.grid(True)
    return ax

def data_bar(xdata, ydata, ax):
    ax.bar(xdata, ydata, color ='green')
    ax.set_xlabel('Месяцы')
    ax.set_ylabel('Количество страниц')
    ax.set_title('Прочитанные страницы за 2024 год')
    ax.legend(title = "Прочитано страниц")
    ax.grid(True)
    return ax


fig, axes = plt.subplots(2, 1)
books_2023 = df[df['Год'] == 2023]
data_plot(books_2023['Месяц'], books_2023['Книги'], axes[0])

pages_2024 = df[df['Год'] == 2024]
data_bar(pages_2024['Месяц'], pages_2024['Страницы'], axes[1])

plt.tight_layout()
plt.show()
```
